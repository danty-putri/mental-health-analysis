## Data Cleaning

- Menghapus kolom duplikat `SocialWeakness.1
- Mengisi missing value pada `SelfEmployed` dengan kategori **‚ÄúUnknown‚Äù**`
- Standardisasi nilai kategorikal (Yes/No/Maybe/Not sure)
- Membersihkan spasi, kapitalisasi, dan kategori yang tidak konsisten

```python
df.isnull().sum()
```

- Kolom **SelfEmployed** memiliki jumlah nilai hilang yang sangat besar.  
- Karena dataset bersifat *self-reported*, kemungkinan besar nilai hilang menunjukkan responden memilih untuk tidak mengungkapkan status pekerjaan mereka.  
- Untuk menjaga integritas data tanpa menghapus banyak baris, nilai hilang akan dikategorikan sebagai **"Unknown"**.


```python
df['SelfEmployed'] = df['SelfEmployed'].fillna("Unknown")
```

```python
df['SocialWeakness'].equals(df['SocialWeakness.1'])
```

```python
df = df.drop(columns=['SocialWeakness.1'])
```

- Dataset memiliki dua kolom dengan nama sama: SocialWeakness dan SocialWeakness.1. Setelah dicek, kedua kolom memiliki nilai yang sama persis, sehingga SocialWeakness.1 merupakan kolom duplikat yang muncul saat proses import data. Untuk menjaga kebersihan dan menghindari redundansi, kolom SocialWeakness.1 dihapus dari dataset.

```python
# Standardisasi nilai kategorikal

def clean_category(value):
    if isinstance(value, str):
        # Hilangkan spasi di awal/akhir
        x = value.strip().lower()

        # Ubah ke format yang konsisten
        if x in ["yes", "y"]:
            return "Yes"
        elif x in ["no", "n"]:
            return "No"
        elif x in ["maybe", "may be"]:
            return "Maybe"
        elif x in ["not sure", "unsure", "not_sure"]:
            return "Not sure"
        else:
            # Capitalize untuk nilai lainnya
            return x.capitalize()
    return value

# Terapkan fungsi ke seluruh DataFrame
df_clean = df.applymap(clean_category)

# Lihat hasil
df_clean.head()
```

#### Standardisasi Nilai Kategorikal

Dataset ini memiliki banyak variabel kategorikal dengan nilai seperti "Yes", "No", "Maybe", dan "Not sure". Namun, perbedaan kecil seperti kapitalisasi atau spasi dapat menyebabkan nilai kategori terpecah menjadi kelas yang berbeda. Oleh sebab itu, kita melakukan standardisasi untuk memastikan format nilai kategorikal konsisten.

Langkah:
- Mengubah seluruh string menjadi huruf kapital awal (title-case)
- Menghapus spasi berlebih
- Menangani nilai "Not sure" menjadi format yang seragam

## 1Ô∏è‚É£ Durasi di Dalam Rumah dan Tingkat Stres

Visualisasi ini bertujuan untuk melihat bagaimana durasi seseorang berada di dalam rumah (`DaysIndoors`)
berhubungan dengan tingkat stress yang dialami (`IncreasingStress`).

```python
# Analisis hubungan DaysIndoors dan IncreasingStress

ct = pd.crosstab(
    df_clean['DaysIndoors'], 
    df_clean['IncreasingStress'],
    normalize='index'
) * 100

ct.round(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gaya visual
sns.set(style="whitegrid", palette="Spectral")

# Buat figure
fig, ax = plt.subplots(figsize=(9,6))

# Plot stacked bar chart dengan warna flare
ct.plot(kind='barh',
        stacked=True,
        ax=ax,
        colormap='Spectral',
        edgecolor='black',
        width=0.75)

# Tambahkan label persentase di setiap segmen batang
for i, (index, row) in enumerate(ct.iterrows()):
    x_offset = 0
    for col in ct.columns:
        value = row[col]
        if value > 0:
            ax.text(x_offset + value/2, i, f'{value:.1f}%', 
                    ha='center', va='center', fontsize=9, color='black')
        x_offset += value

# Judul dan label sumbu
ax.set_title("Proporsi Tingkat Stres Berdasarkan Durasi di Dalam Rumah",
             fontsize=14, fontweight='bold', color='#6A1E55')
ax.set_xlabel("Persentase (%)", fontsize=12)
ax.set_ylabel("Days Indoors", fontsize=12)

# Legend
plt.legend(title="Increasing Stress", bbox_to_anchor=(1.05, 1), loc='upper left',
           frameon=True, fancybox=True, shadow=True)

# Layout rapi
sns.despine()
plt.tight_layout()
plt.show()
```

**Insight:**

Semakin lama seseorang berada di rumah (> 2 bulan), semakin tinggi kemungkinan mengalami stres.
Sedangkan orang yang keluar rumah setiap hari memiliki tingkat stres paling rendah.


## 2Ô∏è‚É£ Pengaruh Riwayat Keluarga terhadap Kesehatan Mental Individu

Visualisasi ini menunjukkan hubungan antara riwayat gangguan mental dalam keluarga (`FamilyHistory`)
dan riwayat gangguan mental pribadi (`MentalHealthHistory`). Tujuannya untuk melihat apakah faktor
genetik memiliki kontribusi terlihat dalam data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Crosstab ulang (kalau belum didefinisikan)
cross_tab2 = pd.crosstab(df["FamilyHistory"], df["MentalHealthHistory"], normalize='index') * 100

# Plot pie chart per kategori FamilyHistory
fig, axes = plt.subplots(1, len(cross_tab2), figsize=(10,4))

for i, col in enumerate(cross_tab2.index):
    axes[i].pie(cross_tab2.loc[col],
                labels=cross_tab2.columns,
                autopct='%1.1f%%',
                startangle=90,
                colors=sns.color_palette('flare'))
    axes[i].set_title(f'Family History: {col}', fontsize=12, fontweight='bold')

plt.suptitle("Distribusi Mental Health History Berdasarkan Family History",
             fontsize=14, fontweight='bold', color='#234E70')
plt.tight_layout()
plt.show()
```

Hasil pie chart menunjukkan **proporsinya relatif mirip** antara responden yang memiliki maupun tidak memiliki riwayat keluarga.

**Insight:**
Dalam dataset ini, **faktor genetik tidak dominan** sebagai penentu utama riwayat kesehatan mental pribadi.

## 3Ô∏è‚É£ Dampak Tingkat Stres terhadap Faktor Psikologis dan Sosial

Visualisasi ini bertujuan memahami dampak stress terhadap faktor psikologis dan sosial


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set tema visual agar tampil lembut dan modern
sns.set_theme(style="whitegrid", palette="Set2")

# Daftar faktor yang akan dibandingkan
factors = ['MoodSwings', 'WorkInterest', 'CopingStruggles', 'SocialWeakness']

# Buat canvas 2x2
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# Warna palet yang serasi
colors = sns.color_palette("Set2")

# Loop untuk setiap faktor
for i, factor in enumerate(factors):
    # Buat tabulasi silang dan ubah ke persentase (%)
    ctab = pd.crosstab(df_clean['IncreasingStress'], df_clean[factor], normalize='index') * 100
    ctab.plot(kind='bar', ax=axes[i], color=colors, edgecolor='black', width=0.75)

    # Judul dan label sumbu
    axes[i].set_title(f"Increasing Stress vs {factor}", fontsize=13, fontweight='bold', color="#333")
    axes[i].set_xlabel("Increasing Stress", fontsize=11)
    axes[i].set_ylabel("Persentase (%)", fontsize=11)
    axes[i].set_ylim(0, 100)
    axes[i].legend(title=factor, loc='upper right')

    # Tambahkan label angka di atas bar
    for p in axes[i].patches:
        height = p.get_height()
        if height > 0:
            axes[i].annotate(f"{height:.1f}%", 
                             (p.get_x() + p.get_width() / 2., height),
                             ha='center', va='bottom',
                             fontsize=9, color='#333', rotation=0)

# Tambahkan judul besar untuk seluruh figure
plt.suptitle("Pengaruh Tingkat Stres terhadap Faktor Psikologis dan Sosial", 
             fontsize=16, fontweight='bold', color="#2F4F4F")

# Tata letak lebih rapi
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

#### üîπ MoodSwings

Tingkat stres tinggi (Yes) = proporsi MoodSwings ‚ÄúHigh‚Äù meningkat signifikan.
‚û° **Stres ‚Üí Emosi tidak stabil.**

#### üîπ WorkInterest

Semakin tinggi stres, semakin banyak responden menjawab **No** pada WorkInterest.
‚û° **Stres menurunkan motivasi dan minat kerja.**

#### üîπ CopingStruggles

Kategori Yes meningkat pada responden dengan IncreasingStress = Yes.
‚û° **Stres mempersulit kemampuan mengatasi tekanan.**

#### üîπ SocialWeakness

Ada peningkatan ringan pada kategori Yes.
‚û° **Pengaruhnya ada, tapi tidak sebesar mood dan coping.**
